```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
of_sim = pd.read_parquet(r"C:\GitRepos\IEAtask49_deepDesign\results\v3\P200_C135_L1430_clump40_fat_0.parquet")

mpy_sim = pd.read_parquet(r"C:\Work_offline\task47\v4\P200_C135_L1430_clump40_fat_0\P200_C135_L1430_clump40_fat_0_meanLineTensions.parquet")
mpy_sim.index = mpy_sim.index + '.outb'
```

```python
motType = ['PtfmSurge', 'PtfmSway', 'PtfmHeave', 'PtfmRoll', 'PtfmPitch', 'PtfmYaw']
of_sim_mot    = of_sim.xs('mean', axis = 1, level = 1).loc[:, motType].copy()
```

```python
common_idx = mpy_sim.index.intersection(of_sim.index)

diff = of_sim_mot.loc[common_idx] - mpy_sim.loc[common_idx]
```

```python
pltQty = ['PtfmSurge', 'PtfmSway']

fig, axs = plt.subplots(figsize=(20, 8), nrows=2, ncols=len(pltQty))

axs[0,0].bar(np.arange(common_idx.shape[0]) + 1, of_sim_mot.loc[common_idx, 'PtfmSurge'], width=1.0, color='blue', label = 'OpenFAST')
axs[0,0].bar(np.arange(common_idx.shape[0]) + 1, mpy_sim.loc[common_idx, 'PtfmSurge'], width=1.0, color='red', alpha=0.5, label = 'Moorpy')
axs[0,0].set_title('PtfmSurge')
axs[0,0].legend()

axs[0,1].bar(np.arange(common_idx.shape[0]) + 1, of_sim_mot.loc[common_idx, 'PtfmSway'], width=1.0, color='blue')
axs[0,1].bar(np.arange(common_idx.shape[0]) + 1, mpy_sim.loc[common_idx, 'PtfmSway'], width=1.0, color='red', alpha=0.5)
axs[0,1].set_title('PtfmSway')

axs[1,0].bar(np.arange(common_idx.shape[0]) + 1, diff.loc[common_idx, 'PtfmSurge'], width=1.0, color='blue')
axs[1,0].set_title('Difference in PtfmSurge')

axs[1,1].bar(np.arange(common_idx.shape[0]) + 1, diff.loc[common_idx, 'PtfmSway'], width=1.0, color='blue')
axs[1,1].set_title('Difference in PtfmSway')

```
